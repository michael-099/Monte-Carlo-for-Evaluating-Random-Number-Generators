### circle

```python
import matplotlib.pyplot as plt
import numpy as np

# Set up the figure and axis
fig, ax = plt.subplots(figsize=(6, 6))

# Draw the square
square_x = [-1, 1, 1, -1, -1]
square_y = [-1, -1, 1, 1, -1]
ax.plot(square_x, square_y, color='blue', label='Square')

# Draw the unit circle
theta = np.linspace(0, 2 * np.pi, 500)
circle_x = np.cos(theta)
circle_y = np.sin(theta)
ax.plot(circle_x, circle_y, color='red', label='Unit Circle')

# Set equal scaling
ax.set_aspect('equal', 'box')

# Add labels and legend
ax.set_title('Unit Circle Inscribed in a Square')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.axhline(0, color='black', linewidth=0.5, linestyle='--')  # X-axis
ax.axvline(0, color='black', linewidth=0.5, linestyle='--')  # Y-axis
ax.legend()

# Show the plot
plt.grid(True)
plt.show()
```
